```python
import sys
sys.path.append("..")
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
relations = pd.read_csv('../Data/CTD/CTD_chemicals_diseases.csv')
```

```python
relations
```

```python
relations = relations[['ChemicalID', 'DiseaseID', 'DirectEvidence', 'InferenceScore']]
```

```python
relations
```

```python
indexs = []
for index, row in tqdm(relations.iterrows()):
    indexs.append(index)
    relations.at[index,'DiseaseID'] = row['DiseaseID']
```

```python
len(indexs)
```

```python
relations = relations.loc[indexs]
```

```python
relations['InferenceScore'].fillna(value=relations['InferenceScore'].mean(), inplace=True)
```

```python
relations
```

```python
bool_series = relations.duplicated(keep='first')
relations = relations[~bool_series]
```

```python
relations
```

```python
relations.to_csv('../Data/CTD/CTD_chemicals_diseases_result.csv')
```

```python
relations = pd.read_csv('../Data/CTD/CTD_genes_diseases.csv', skiprows=27)
```

```python
len(relations)
```

```python
relations
```

```python
relations = relations[['# GeneSymbol', 'GeneID', 'DiseaseID']]
```

```python
relations = relations[1:]
```

```python
relations
```

```python
bool_series = relations.duplicated(keep='first')
relations = relations[~bool_series]
```

```python
relations
```

```python
relations.to_csv('../Data/CTD/CTD_genes_diseases_result.csv', index=False)
```

```python
dis_gens = dict()
for index, row in tqdm(relations.iterrows()):
    gene_id = row['GeneID']
    gene_name = row['# GeneSymbol']
    disease_infor = row['DiseaseID']
    if disease_infor not in dis_gens:
        dis_gens[disease_infor] = list()
    if gene_name not in dis_gens[disease_infor]:
        dis_gens[disease_infor].append(gene_name)
```

```python
from json import JSONEncoder
import os
import json
```

```python
class ObjectEncoder(JSONEncoder):
    def default(self, o):
        return o.__dict__
```

```python
with open('../Data/CTD/CTD_diseases_genes_dict.json', 'w+') as file:
    file.write(json.dumps(dis_gens, cls=ObjectEncoder, indent=2, sort_keys=True))
```
