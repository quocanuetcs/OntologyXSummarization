```python
import xml.etree.cElementTree as ET
from tqdm import tqdm
import pickle
```

```python
with open('Data/MeSH/mtrees2021.bin', mode='rb') as file: 
    content = file.read()
```

```python
IDs = []
for nodestr in str(content).split('\\n')[:-1]:
    try:
        ID = nodestr.split(';')[1]
        IDs.append(ID)
    except:
        print(nodestr)
```

```python
def get_parent(ID):
    parentID = ID.split('.')
    return '.'.join(parentID[:-1])
```

```python
parent_dict = dict()
```

```python
for ID in IDs:
    parentID = get_parent(ID)
    if parentID not in IDs:
        print(ID, parentID)
    else:
        if parentID not in parent_dict:
            parent_dict[parentID] = list()
        parent_dict[parentID].append(ID)
```

```python
with open('Data/MeSH/mtrees2021.pkl', 'wb') as f:
    pickle.dump(parent_dict, f)
```
