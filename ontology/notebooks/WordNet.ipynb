```python
def synonym_antonym_extractor(phrase):
    from nltk.corpus import wordnet
    synonyms = []
    antonyms = []

    for syn in wordnet.synsets(phrase):
        for l in syn.lemmas():
            synonyms.append(l.name())
            if l.antonyms():
                antonyms.append(l.antonyms()[0].name())

    print(set(synonyms))
    print(set(antonyms))

synonym_antonym_extractor(phrase="treat")
```

```python
synonym_antonym_extractor(phrase="treatment")
```

```python
s = 'treat'
word_list = wn.words(s)
```

```python
for w in word_list:
#     print('word:', word)
#     print('form:', word.forms())
#     print('lemma:', word.lemma())
#     print('derived_words:', word.derived_words())
#     print('synsets:', word.synsets())
    for synset in w.synsets():
        print('synset', synset)
        print(synset.words())
        for word in synset.words():
            print('word:', word)
            print('form:', word.forms())
            print('lemma:', word.lemma())
            print('derived_words:', word.derived_words())
            for drived_word in word.derived_words():
                print('-drived_word:', drived_word)
                print('-drived_word form:', drived_word.forms())
                print('-drived_word lemma:', drived_word.lemma())
            print(synset.lemmas())
            print('------------')
    print()
```
