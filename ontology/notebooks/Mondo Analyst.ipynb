```python
import sys
sys.path.append("..")
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
data = pd.read_csv('../Data/Mondo/MONDO.csv')
```

```python
for val in data["Parents"].values:
    print(val)
```

```python
def get_id(s):
    try:
        rs = s.split('/')[-1]
    except:
        rs = "None"
    return rs
```

```python
for idx in tqdm(data.index):
    row = data.iloc[idx]
    ID = get_id(row["Class ID"])
    parID = get_id(row["Parents"])
    data.loc[idx, "Class ID"] = ID
    data.loc[idx, "Parents"] = parID
```
